```python
headers = {"User-Agent": "Mozilla/5.0 (X11; Linux i686) AppleWebKit/537.17 (KHTML, like Gecko) Chrome/24.0.1312.27 Safari/537.17"}
```

```python
def get_request(req_type: str, item_id: int):
    '''
    Function returns the request url
    
    example: https://api.themoviedb.org/3/movie/603?api_key=fa0b74631406cb7e556e50abc1edb711&language=en-US
    
    @param req_type: movie or person 
    @item_id: ID number
    @returns: request URL 
    '''
    base_url = 'http://api.themoviedb.org/3/'
    api_key = 'fa0b74631406cb7e556e50abc1edb711'
    
    return (base_url + req_type + '/' + str(item_id) + '?api_key=' + api_key+'&language=en-US')
    
```

```python
#Example to get The Matrix with function 
req_matrix = get_request('movie', 603)
```

```python
print(req_matrix)
```

```python
matrix_obj = urllib.request.urlopen(req_matrix)
```

```python
matrix_data = matrix_obj.read()
```

```python
encoded_matrix = matrix_obj.info().get_content_charset('utf8')
```

```python
matrix_dict = json.loads(matrix_data.decode(encoded_matrix))
```

```python
matrix_dict
```

So the process to obtain the dictionary: 

1. Create HTTPConnect object 
2. Get data by reading HTTPConnect object 
3. encode HTTPConnect object 
4. Use data to decode and pass through json.loads()
5. A list of dictionaries is returned
